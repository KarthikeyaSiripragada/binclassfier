```python
# 3) securely write ~/.netrc with your token (interactive)
import getpass, os, stat, textwrap, sys

GITHUB_USER = "YOUR_GITHUB_USERNAME"   # <-- replace
GITHUB_REPO = "SmartBin"               # <-- replace if needed

token = getpass.getpass("Enter your GitHub Personal Access Token (classic, repo scope): ")

netrc_content = textwrap.dedent(f"""
    machine github.com
      login {GITHUB_USER}
      password {token}
""").strip() + "\n"

netrc_path = os.path.expanduser("~/.netrc")
with open(netrc_path, "w") as f:
    f.write(netrc_content)
os.chmod(netrc_path, 0o600)
print("Wrote ~/.netrc (mode 600).")
```
